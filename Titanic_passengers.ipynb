# Titanic Data Analysis for Decision Making

## Objective
The goal of this project is to analyze passenger data and identify patterns that affect survival, in order to demonstrate how data analysis can support decision-making.

## Tools
- Python (Pandas, Matplotlib)

## Dataset
Titanic dataset (public dataset)

## Data Loading

In this section, we load the dataset and explore its structure.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

df.head()
```

## Data Overview

We check the data types and missing values in the dataset.

```python
df.info()
```

## Survival Analysis

We analyze how many passengers survived.

```python
df['Survived'].value_counts()
```

```python
df['Age'].mean()
```

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Count")
plt.show()
```

## Survival by Gender

We analyze how survival rates differ between male and female passengers.

```python
df.groupby('Sex')['Survived'].mean()
```

```python
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Gender")
plt.show()
```

## Survival by Passenger Class

We analyze how passenger class affects survival rates.

```python
df.groupby('Pclass')['Survived'].mean()
```

```python
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')
plt.title("Survival Rate by Passenger Class")
plt.show()
```

## Age Distribution

We explore the distribution of passenger ages.

```python
df['Age'].plot(kind='hist', bins=20)
plt.title("Age Distribution")
plt.show()
```

## Key Insights

- Female passengers had a significantly higher survival rate than males.
- Passengers in higher classes (1st class) had better chances of survival.
- Age distribution shows a wide range of passengers, with some missing values.

## Conclusion

This analysis demonstrates how data can reveal important patterns and support data-driven decision-making. Factors such as gender and passenger class had a strong impact on survival outcomes.
